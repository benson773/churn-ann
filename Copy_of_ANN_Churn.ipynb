```python
# ================================
# ANN for Bank Customer Churn (Colab-ready version)
# With clear, student-style comments
# ================================

# Step 0: imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
# In Colab, you'll upload the CSV from your computer
from google.colab import files
uploaded = files.upload()   # pick your CSV file (e.g. Churn_Modelling (1).csv)

# Automatically get the uploaded file name
filename = list(uploaded.keys())[0]
print("Uploaded file:", filename)

# Read CSV
dataset = pd.read_csv(filename)

# Quick checks
print("Dataset shape:", dataset.shape)
display(dataset.head())      # show first 5 rows
print(dataset.info())

# Step 2: Separate features (X) and target (y)
X_df = dataset.iloc[:, 3:13].copy()   # take columns 3 to 12
y = dataset.iloc[:, 13].values        # take column 13 (Exited)

print("\nFeature columns before encoding:", list(X_df.columns))
print("X_df shape:", X_df.shape, " y shape:", y.shape)

# Step 3: Preprocessing - encode categorical variables
# Geography -> one-hot encode (drop_first=True removes one dummy)
# Gender -> map to 0/1

if 'Geography' not in X_df.columns or 'Gender' not in X_df.columns:
    raise ValueError("Expected 'Geography' and 'Gender' columns in dataset. Check CSV headers.")

X_df = pd.get_dummies(X_df, columns=['Geography'], drop_first=True)
X_df['Gender'] = X_df['Gender'].map({'Female': 0, 'Male': 1})

if X_df['Gender'].isnull().any():
    raise ValueError("Unexpected values in 'Gender' column. Check spelling in CSV.")

print("\nFeature columns after encoding:", list(X_df.columns))
print("X_df shape (after encoding):", X_df.shape)

# Convert to numpy for model
X = X_df.values.astype(np.float64)

# Step 4: Split and scale data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print("\nShapes after split & scaling:")
print("X_train:", X_train.shape, "X_test:", X_test.shape)

# Step 5: Build ANN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(units=6, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.1),
    Dense(units=6, activation='relu'),
    Dropout(0.1),
    Dense(units=1, activation='sigmoid')   # sigmoid = binary output
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Step 6: Train model
history = model.fit(X_train, y_train, batch_size=10, epochs=50, validation_split=0.1)

# Plot training history
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend(); plt.title('Loss')

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend(); plt.title('Accuracy')
plt.show()

# Step 7: Evaluate on test set
y_prob = model.predict(X_test).ravel()
y_pred = (y_prob > 0.5).astype(int)

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 8: Single-sample prediction
print("\nFeature order used for model:\n", list(X_df.columns))

# Example input (change values to test different customers)
sample_dict = {
    'CreditScore': 600,
    'Gender': 1,           # Male=1, Female=0
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000,
    # Geography dummies like 'Geography_Germany' or 'Geography_Spain' must be included
}

# Build sample row
sample = np.zeros(len(X_df.columns), dtype=float)
cols = list(X_df.columns)
for i, col in enumerate(cols):
    if col in sample_dict:
        sample[i] = sample_dict[col]
    else:
        sample[i] = 0.0

# Scale and predict
sample_scaled = sc.transform(sample.reshape(1, -1))
prob = model.predict(sample_scaled)[0][0]
print("\nSingle sample churn probability:", prob)
print("Predicted class (1=churn):", int(prob > 0.5))
```
